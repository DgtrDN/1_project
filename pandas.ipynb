```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Создадим множества
age_missing = set(train[train['Age'].isnull()].index)
cabin_missing = set(train[train['Cabin'].isnull()].index)

# Построим диаграмму Венна
plt.figure(figsize=(8, 6))
venn2([age_missing, cabin_missing], ('Пропуски Age', 'Пропуски Cabin'))
plt.title('Пересечение пропусков Age и Cabin')
plt.show()
```
