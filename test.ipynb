```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
data = pd.read_csv('amazon_sales_dataset.csv')
# Display the first few rows of the dataset
data.head(5)
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
plt.scatter(data['price'], data['quantity_sold'], color='blue', label='Data Points', alpha=0.6)
plt.title('Scatter Plot of Price vs Quantity Sold')
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.legend()
plt.grid()
plt.show()
```
